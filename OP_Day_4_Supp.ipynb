```python
import numpy as np
import random
from matplotlib import pyplot as plt
```

```python
x = 10*np.random.rand(100000)
y = 3*x + 3*np.random.randn(100000)
r = random.sample(range(100000),30)
plt.scatter(x[r],y[r])
```

```python
def MSE_loss(theta,x,y):
    return np.linalg.norm(theta*x-y)**2
```

```python
def D_loss(theta,x,y):
    return 2*x*(theta*x-y)
```

```python
Loss = []
theta_list = []
theta = np.random.rand() # theta 초기값
epoch = 50
alpha = 0.01
batch = 20
for _ in range(epoch):
    Loss.append(MSE_loss(theta,x,y))
    theta_list.append(theta)
    t_d = 0
    for __ in range(batch):
        n = np.random.randint(1000)
        t_d = t_d + D_loss(theta,x[n],y[n])
    t_d = t_d/20
    theta -= 0.01*t_d
```

```python
plt.plot(Loss)
```
